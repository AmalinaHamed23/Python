```python
#TYPE CASTING
#IT'S A PROCESS OF CONVERTING ONE DATA TYPE TO ANOTHER DATA TYPE

#automatic type convertion (IMPLICIT)
#lower data type will be converted to higher data type automatically

#highest data type (higher priority,result will show higher priority data type)= string-- float --integer

a = 10
b = 45.03
res = a+b  #10.00+45.03
print(res)
```

```python
#manual type conversion (EXPLICIT)
#PERFORMED BY DEVELOPER

a = 10
b = 45.03
res = a + int(b)  #10 + 45
print(res)

a = 10
b = 45.97
res = a + int(b)
print(res)

# manual type conversion  EXPLICIT
# performed by developers

a = 10
b = 45.03
res = a + int(b)   # 10 + 45
print(res)


# we have a floatVar which needs to be converted into intVar
#  int(floatvar)      int(23.33) --> 23

# we have an intVar which needs to be converted into floatVar
#  float(intVar)     float(10)  --> 10.0

# we have a intVar which needs to be converted into strVar
#  str(intVar)         str(24)   --> '24'

# we have a floatVar which needs to be converted into strVar
#  str(floatVar)         str(24.334)   --> '24.334'

# converting bool into int and float

choice = True
print(int(choice))
print(float(choice))

choice = False
print(int(choice))
print(float(choice))

print(str(choice))

res = 1
print(bool(res))

res = 0
print(bool(res))

res = 76
print(bool(res))

#name = 'John'
#print(int(name))   CANNOT *see notebook
```

```python
#OPERATORS
#operator are used to perform operations on variables

# a.) ARITHMETIC OPERATORS
# are used to perform mathematical operations on variables
#  +  -  *  /  //  **

print(10+5.6)
print(10-5)
print(10*5)
print(10/5)  #IN PYTHON DIVISION OPERATOR / USED TO RETURN VALUE IN FLOAT
print(10//5) #IN PYTHON FLOOR DIVISION / USED TO RETURN VALUE IN INT
print(10%3)   #MODULUS USED TO GIVE REMAINDER
print(2**8)   # 2 RAISE TO THE POWER OF 8    **exponent
print(72**72)
```

```python
# c) COMPARISON OPERATORS
# <  >  ==  !=  <=  >=
# USED TO RETURN A BOOLEAN VALUE [TRUE/FALSE]

print(10<5)
print(10>5)
print(98==98)
print(100!=1000)
print(100!=100)

age = 21
print(age>=18)  #age should be equal to 18 or age should be greater than 18
age = 18
print(age>=18)

# real life example : taxi drive must be 60 or below, weight limit for bungee jumping (must be 70kg or below)
```

```python
# D) LOGICAL OPERATORS
# works with multiple conditions
# and or not
# USED TO RETURN A BOOLEAN VALUE [TRUE/FALSE]

# and
# and requires more than one condition
# and will return TRUE when both or all the conditions are TRUE

print(100==100 and 200==200)
 #       True          True

print(100==100 and 200==2000)
 #       True          False

print(100==100 and 200==200 and 300==350)
 #       True          True       False

 # real life example - login gmail, username and pw must be correct, condition for playing game, must be 18 above and malaysian
```

```python
# or
# or requires more than one condition
# or will return TRUE when both or all the conditions are TRUE
# or will also returns True if atleast any condition is TRUE
# or will return False when both or all the conditions are FALSE

print(100==100 or 200==300)
 #       True          False

print(100==150 or 200==2000)
 #     FAlse          False

print(100==100 or 200==200 and 300==350)
 #       True          True       False

 # real life example : can make payment by cash or card. can choose dining or take out. who can take driver license, m or f. as a game developer, who can download the game, malaysian or singapore or japan or india
```

```python
# not  *rarely used
# used to reverse the result

print(not(100==100 and 200==200))

# real life example : banking application automatically logout if user is not active for 10 mins
                      # session login : True  --> not active for 10 mins --> not session --> logout
```
